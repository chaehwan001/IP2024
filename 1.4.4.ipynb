##### chap 1.4.4 Image Blurring (Image Smoothing)


#### Averaging

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('winter.jpg')
blur3 = cv2.blur(img,(3,3))
blur5 = cv2.blur(img,(5,5))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur5),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

####  Gaussian Filtering

```python
img = cv2.imread('winter.jpg')
blur = cv2.blur(img,(5,5))
blur5 = cv2.GaussianBlur(img,(5,5),0)
plt.subplot(121),plt.imshow(blur),plt.title('blur')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur5),plt.title('Gaussian Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

#### Median Filtering

```python
img = cv2.imread('winter.jpg')
#blur = cv2.blur(img,(5,5))
blur5 = cv2.GaussianBlur(img,(5,5),0)
medianBlur = cv2.medianBlur(img,5)

plt.subplot(121),plt.imshow(blur5),plt.title('blur5')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(medianBlur),plt.title('medianBlur')
plt.xticks([]), plt.yticks([])
plt.show()
```

#### Bilateral Filtering

```python
img = cv2.imread('winter.jpg')
#blur = cv2.blur(img,(5,5))
blur5 = cv2.GaussianBlur(img,(5,5),0)
medianBlur = cv2.medianBlur(img,5)
Bilateralblur5 = cv2.bilateralFilter(img,9,75,75)

plt.subplot(121),plt.imshow(blur5),plt.title('blur5')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(Bilateralblur5),plt.title('Bilateralblur5')
plt.xticks([]), plt.yticks([])
plt.show()
```
